# Fixed versus variable costs (EOQ)


* D = the demand rate of the item, in units / unit time.
* A = the fixed cost component.
* v = the unit variable cost of the item, in money / unit.
* r = the carrying charge, the cost of having one dollar of the item tied up in the inventory for a unit time interval, in money / money / unit.
* Q = the replenishment order quantity, in units.
* T = the time between placement (and receipt) of replenishment orders.

```python
%matplotlib inline
eoq_1 = EOQ(D=12000, A=20.0, v=1.0, r=0.1)
eoq_1.compute()
eoq_1.print()
eoq_1.plot()
```
